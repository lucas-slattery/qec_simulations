```python
def cum_est_fid(discarded_sv_norms,norms):
    cum_est_fids = []
    fid = 1
    for i,(d_sv,norm) in enumerate(zip(discarded_sv_norms,norms)):
        fid *= (1-d_sv/norm)**2
        cum_est_fids.append(fid)

    return cum_est_fids
```

```python
params = bb_code_parameters[0]
bb_code_results = collated_results[str(params)]

cm = colormaps['viridis']
num_ancillas = 1
chi = 256
for i, params in enumerate(bb_code_parameters):
    bb_code_results = collated_results[str(params)]

    color = cm(i/(len(constraints)-1))
    for use_lazy_indexing in ['true','false']:
        ls = '-' if use_lazy_indexing == 'true' else '-.'
    
        norms = bb_code_results[str(num_ancillas)][use_lazy_indexing][str(chi)]['norms']
        dsvs = bb_code_results[str(num_ancillas)][use_lazy_indexing][str(chi)]['discarded_sv_norms']

        cum_fids = cum_est_fid(dsvs,norms)

        xs = np.linspace(0,100,len(cum_fids))
        plt.plot(xs, np.array(cum_fids), color = color, ls = ls ,label = f"use_lazy_indexing = {use_lazy_indexing}, BB Code = {params}")

plt.yscale('log')
plt.ylabel(r'$\mathcal{F}$')
plt.xlabel('% of Simulation')
plt.legend()
plt.title(f'BB Code with # of Ancillas = {num_ancillas}')
plt.savefig(plots_folder + f"bb_code_cum_fid.png")
```

```python
bb_code_parameters = StabilizerCodeFactory.build_bivariate_bicycle_code_valid_parameters()[0:13:4]

cm = colormaps['viridis']
num_ancillas = 1
use_lazy_indexing = 'true'
plt.figure(figsize=(10,10))
for i, params in enumerate(bb_code_parameters):
    

    color = cm(i/len(constraints))
    for use_gpu in ['true','false']:
        if use_gpu == 'true':
            bb_code_results = gpu_collated_results[str(params)]
        else:
            bb_code_results = cpu_collated_results[str(params)]
        ls = '-' if use_gpu == 'true' else '-.'
    
        chis = np.array([1,2,4,8,16,32,64,128,256,512,1024])
        times = []
        fids = []
        for chi in chis:
            try:
                time = bb_code_results[str(num_ancillas)][use_lazy_indexing][str(chi)]['time']
                fid = bb_code_results[str(num_ancillas)][use_lazy_indexing][str(chi)]['estimated_fidelity']
            except:
                time = 0
                fid = 0
            times.append(time)
            fids.append(fid)

        times = np.array(times)
        fids = np.array(fids)
        
        ys = times[times > 0]
        xs = chis[times > 0]

        plt.scatter(xs,ys, color = color,)
        plt.plot(xs,ys, color = color, ls = ls,label = f"BB Code = {params}, use_gpu = {use_gpu}")
        
plt.yscale('log')
plt.xscale('log')
plt.ylabel('Time (s)')
plt.xlabel(r'$\chi$')
plt.grid(which='both')
plt.legend()
plt.title(f'BB Code with # of Ancillas = {num_ancillas}')
plt.savefig(plots_folder + f"bb_code_time_vs_chi.png")
```
